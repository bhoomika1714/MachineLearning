```python
# Define the rescue calculator function
def rescue_calculator(detections, rescue_capacity=2, boat_capacity=4):
    """
    Calculate the number of rescue personnel and boats required based on the number of detected persons and animals.

    Parameters:
    detections (dict): Dictionary with counts of detected objects, such as {'person': 5, 'animal': 3}.
    rescue_capacity (int): Number of people one rescue person can handle (default is 2).
    boat_capacity (int): Number of people/animals one boat can handle (default is 4).

    Returns:
    dict: Dictionary with the number of rescue personnel and boats needed.
    """
    # Initialize counts for people and animals
    people_count = detections.get('person', 0)
    animal_count = detections.get('animal', 0)
    total_rescue_needed = people_count + animal_count

    # Edge case: if no persons or animals are detected
    if total_rescue_needed == 0:
        return {
            "rescue_personnel": 0,
            "boats": 0,
            "message": "No people or animals detected. No rescue needed."
        }

    # Calculate required rescue personnel
    rescue_personnel_needed = (total_rescue_needed + rescue_capacity - 1) // rescue_capacity  # Round up if needed

    # Calculate required boats
    boats_needed = (total_rescue_needed + boat_capacity - 1) // boat_capacity  # Round up if needed

    # Output the result
    return {
        "rescue_personnel": rescue_personnel_needed,
        "boats": boats_needed,
        "total_rescue_needed": total_rescue_needed
    }

# Example usage of the function with sample detection data
sample_detections = {'person': 5, 'animal': 3}
rescue_calculator(sample_detections)
```
